```python
import numpy as np
sns.gender.replace({'F':0,'M':1},inplace=True)
sns['age'].fillna(np.mean(sns['age']),inplace=True)
```

```python
from sklearn.cluster import KMeans
m1=KMeans(n_clusters=5)
m1.fit(sns)
```

```python
m1.score(sns)
```

```python
m1.cluster_centers_
```

```python
pd.DataFrame(m1.cluster_centers_)
```

```python
m1_score=[]
for i in range(1,11):
 m1=KMeans(n_clusters=i)
 m1.fit(sns)
 m1_score.append(m1.score(sns)*-1)
print(m1_score)
```

```python
m1_score
```

```python
import matplotlib.pyplot as plt
cluster=[i for i in range(1,11)]
plt.plot(cluster,m1_score,marker="*")
plt.grid()
plt.show()
```
